```python
import pandas as pd
import os

# assign csv dates
datef1 = '010517'
datef2 = '091720'
```

```python
# Read CSV containing cities weather data 1
file = os.path.join('data', f'wx_cities_{datef1}.csv')
df_wx1 = pd.read_csv(file)
print(f"There are {len(df_wx1)} records retrieved from csvfile.")
df_wx1
```

```python
# rearrange columns and rename for consistency table design
df_wx1 = df_wx1[['City_ID','City','Country','Lat','Lng','Date','Cloudiness','Humidity','Max Temp','Wind Speed']]
df_wx1 = df_wx1.rename(columns={'Lng': 'Lon','Date': 'Date (UTC)', 'Cloudiness': 'Clouds (%)', 
                       'Humidity':'Humidity (%)', 'Max Temp': 'Max Temp (F)', 'Wind Speed': 'Wind Speed (mph)'})
df_wx1
```

```python
# Read CSV containing cities weather data 2
file = os.path.join('data', f'wx_cities_{datef2}.csv')
df_wx2 = pd.read_csv(file)
df_wx2 = df_wx2.drop(columns=['Unnamed: 0'])
print(f"There are {len(df_wx2)} records retrieved from csvfile.")
df_wx2
```

```python
#export dataframea to html file
df_wx1.to_html(f'visualizations/{datef1}/weather_data_export.html', index=False)
```

```python
#export dataframea to html file
df_wx2.to_html(f'visualizations/{datef2}/weather_data_export.html', index=False)
```
